这一部分将让你开始思考算法设计和分析。它旨在温和地介绍如何指定算法、在整本书中使用的一些设计策略、以及在算法分析中使用到的许多基本思想。本书往后的部分建立在这个基础之上。

在第1章里，我们概述了算法及其在现代计算系统中的地位。这一章定义了什么是算法，并列出了一些示例。它也说明：**我们应该把算法看做一种技术**，与快速硬件、图形用户界面、面向对象系统、网络等技术并列。

第2章里，我们介绍了第一批算法，它们求解了对n个数字序列的排序问题。它们是用伪代码来书写的，虽然不能直接地转换成任一编程语言，但是伪代码足够清楚地表达了算法的结构，你应该能用自己选择的编程语言来实现它。我们的检查的排序算法是插入排序、归并排序。虽然每个算法要求的时间都随着n增长，但是两个算法之间的增长速度是不同的。我们在第2章里确定这些运行时间，并开发了一种有用的符号来表示它们。

在第3章里，我们定义了这种符号，即渐近符号。这一章通过定义若干个渐近符号开始，我们使用它们来从上面且/或下面来界定算法运行时间。第3章的剩余部分主要是呈现数学符号，更多地是为了保证你使用的数学符号与本书中保持一致，而不是教你新的数学概念。
