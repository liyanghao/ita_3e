这一部分将让你开始思考算法设计和分析。它旨在温和地介绍如何指定算法、在整本书中使用的一些设计策略、以及在算法分析中使用到的许多基本思想。本书往后的部分建立在这个基础之上。

在第1章里，我们概述了算法及其在现代计算系统中的地位。这一章定义了什么是算法，并列出了一些示例。它也说明：**我们应该把算法看做一种技术**，与快速硬件、图形用户界面、面向对象系统、网络等技术并列。

第2章里，我们介绍了第一批算法，它们求解了对n个数字序列的排序问题。它们是用伪代码来书写的，虽然不能直接地转换成任一编程语言，但是伪代码足够清楚地表达了算法的结构，你应该能用自己选择的编程语言来实现它。我们的检查的排序算法是使用**增量方法**的插入排序、使用“分而治之”这一递归技术的归并排序。虽然每个算法要求的时间都随着n增长，但是两个算法之间的增长速度是不同的。我们在第2章里确定这些运行时间，并开发了一种有用的符号来表示它们。

在第3章里，我们定义了这种符号，即渐近符号。这一章通过定义若干个渐近符号开始，我们使用它们来从上面且/或下面来界定算法运行时间。第3章的剩余部分主要是呈现数学符号，更多地是为了保证你使用的数学符号与本书中保持一致，而不是教你新的数学概念。

在第4章里，我们进一步地探讨了在第2章里介绍的分而治之方法。这一章提供了更多分而治之算法的示例，包括`Strassen`的令人惊讶的两个方针相乘方法。这一章还包含了求解递归式的方法，递归式对描述递归算法的运行时间是有用的。一种功能强大的技术是“**主方法**”，我们经常使用它来求解从分而治之算法中产生的递归式。虽然第4章的大部分内容是致力于证明主方法的正确性，但是跳过这个证明，你依然可以使用主方法。

在第5章里，我们介绍了概率分析和随机算法。我们通常使用概率分析来确定一个有如下特征的算法的运行时间：由于固有的概率分布，算法的运行时间可能在相同大小的不同输入上有所不同。有些情况下，我们假设输入服从已知的概率分布，以便我们可对所有的可能输入计算平均运行时间。其他情况下，概率分布不是来自于输入，而是在算法执行过程中的随机选择。一个随机算法的行为由它的输入和随机数生成器产生的值来决定。我们可以给输入强制一个概率分布来使用随机计算法，从而确保没有特定的输入能导致差的性能，甚至可以给算法的错误率设置设置一个界，从而允许算法在有限的基础上产生错误的结果。

在附录A-D里，我们包含了你在阅读本书时会发现有帮助的其他的数学材料。在阅读这本书之前，你可能已经看到过附录章节里的许多材料，所以你应该考虑把附录当做参考资料。另一方面，你可能还没有看到过在本书第一部分里的许多材料。在第一部分里和附录里的所有章节都是用教程的风格来书写的。
